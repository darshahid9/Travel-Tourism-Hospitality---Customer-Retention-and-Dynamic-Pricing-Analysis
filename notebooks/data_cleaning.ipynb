```python
# Import libraries
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset
df = pd.read_csv('hotel_bookings.csv')
print(f'Shape: {df.shape}')
df.head()
```

```python
# Check column data types
df.dtypes
```

```python
# Basic statistics
df.describe()
```

```python
# Remove duplicate rows
before = len(df)
df.drop_duplicates(inplace=True)
print(f'Duplicates removed: {before - len(df)}')
print(f'Rows remaining: {len(df)}')
```

```python
# Check missing values
df.isnull().sum()
```

```python
# Fix missing values
# agent / company: NaN means no agent or company used
df['agent']   = df['agent'].fillna(0).astype(float).astype(int)
df['company'] = df['company'].fillna(0).astype(float).astype(int)

# country: fill with most frequent country
df['country'] = df['country'].fillna(df['country'].mode()[0])

# children: only 4 missing, fill with 0
df['children'] = df['children'].fillna(0).astype(float).astype(int)

print(f'Missing values remaining: {df.isnull().sum().sum()}')
```

```python
# Remove rows with 0 guests (invalid bookings)
before = len(df)
df = df[~((df['adults'] == 0) & (df['children'] == 0) & (df['babies'] == 0))]
print(f'Zero-guest rows removed: {before - len(df)}')
```

```python
# Convert arrival month name to number
month_map = {
    'January':1, 'February':2, 'March':3, 'April':4,
    'May':5, 'June':6, 'July':7, 'August':8,
    'September':9, 'October':10, 'November':11, 'December':12
}
df['arrival_date_month_num'] = df['arrival_date_month'].map(month_map)

# Build proper arrival_date datetime column
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month_num'].astype(str) + '-' +
    df['arrival_date_day_of_month'].astype(str),
    errors='coerce'
)

# Convert reservation status date
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'], errors='coerce')

df[['arrival_date', 'reservation_status_date']].dtypes
```

```python
# Treat ADR outliers using IQR Winsorization (cap instead of drop)
Q1  = df['adr'].quantile(0.25)
Q3  = df['adr'].quantile(0.75)
IQR = Q3 - Q1
lower = max(0, Q1 - 1.5 * IQR)
upper = Q3 + 1.5 * IQR

df['adr'] = df['adr'].clip(lower=lower, upper=upper)
print(f'ADR capped to [{lower:.2f}, {upper:.2f}]')
df['adr'].describe()
```

```python
# Feature Engineering

# Total length of stay
df['total_nights'] = df['stays_in_weekend_nights'] + df['stays_in_week_nights']

# Remove zero-night bookings
df = df[df['total_nights'] > 0]

# Total guests
df['total_guests'] = df['adults'] + df['children'] + df['babies']

# Revenue proxy
df['total_revenue'] = df['adr'] * df['total_nights']

# Season from arrival month
season_map = {
    1:'Winter', 2:'Winter', 12:'Winter',
    3:'Spring', 4:'Spring', 5:'Spring',
    6:'Summer', 7:'Summer', 8:'Summer',
    9:'Autumn', 10:'Autumn', 11:'Autumn'
}
df['season'] = df['arrival_date_month_num'].map(season_map)

# Lead time buckets
def lead_bucket(d):
    if d == 0:      return 'Same-day'
    elif d <= 7:    return 'Last-minute (1-7d)'
    elif d <= 30:   return 'Short (8-30d)'
    elif d <= 90:   return 'Medium (1-3m)'
    elif d <= 180:  return 'Long (3-6m)'
    else:           return 'Very long (6m+)'

df['lead_time_bucket'] = df['lead_time'].apply(lead_bucket)

# Traveler type
df['traveler_type'] = df['market_segment'].apply(
    lambda x: 'Corporate' if x in ['Corporate', 'Aviation'] else 'Leisure'
)

# Binary flags
df['has_special_requests'] = (df['total_of_special_requests'] > 0).astype(int)
df['includes_weekend']     = (df['stays_in_weekend_nights'] > 0).astype(int)

print('Feature engineering done.')
df[['total_nights', 'total_guests', 'total_revenue', 'season', 'lead_time_bucket', 'traveler_type']].head()
```

```python
# Final summary
print(f'Final shape:        {df.shape}')
print(f'Missing values:     {df.isnull().sum().sum()}')
print(f'Cancellation rate:  {df["is_canceled"].mean()*100:.1f}%')
print(f'Avg ADR:            {df["adr"].mean():.2f}')
print(f'Avg stay:           {df["total_nights"].mean():.1f} nights')
print(f'Avg revenue:        {df["total_revenue"].mean():.2f}')
```

```python
# Export clean dataset
df.to_csv('../data/hotel_bookings_clean.csv', index=False)
print('Saved: ../data/hotel_bookings_clean.csv')
```
